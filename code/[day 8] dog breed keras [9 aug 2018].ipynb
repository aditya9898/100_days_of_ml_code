```python
import pandas as pd
import numpy as np
import os
import shutil
```

```python
import tensorflow as tf
import keras
import matplotlib.pyplot as plt
from keras.layers import Dense,GlobalAveragePooling2D
```

```python
from keras.applications import MobileNet
from keras.preprocessing import image
from keras.applications.mobilenet import preprocess_input
from keras.preprocessing.image import ImageDataGenerator
```

```python
from keras.models import Model
from keras.optimizers import Adam
```

```python
df=pd.read_csv('./Desktop/dog breed dataset/labels.csv')
print(df.shape)
labels=df.as_matrix()
```

```python
folder_path='./Desktop/dog breed dataset/train/'
main_new_path='./Desktop/train/'
directory=os.fsencode(folder_path)
for file in os.listdir(directory):
    file_name=os.fsdecode(file)
    code=file_name[0:-4]
    index_of_breed=np.where(labels[:,0]==code)
    name_of_breed=labels[index_of_breed,1]
    print(name_of_breed[0][0])
    new_path=str(main_new_path+name_of_breed[0][0])
    new_path_d=os.fsencode(new_path)
    if not os.path.exists(new_path_d):
        os.makedirs(new_path_d)
    olddir=os.fsencode(str(folder_path+file_name))
    newdir=os.fsencode(str(new_path))
    shutil.copy(olddir,newdir)
```

```python
base_model=MobileNet(weights='imagenet')
```

```python
for i,layer in enumerate(base_model.layers):
    print(i,layer.name)
```

```python
x=base_model.get_layer('conv_pw_13_relu').output
x=GlobalAveragePooling2D()(x)
x=Dense(1024,activation='relu')(x)
x=Dense(1024,activation='relu')(x)
x=Dense(512,activation='relu')(x)
preds=Dense(120,activation='softmax')(x)
```

```python
model=Model(inputs=base_model.input,outputs=preds)
```

```python
for layer in model.layers[:50]:
    layer.trainable=False
```

```python
model.compile(optimizer='Adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
train_datagen=ImageDataGenerator(samplewise_std_normalization=True,
                                 shear_range=0.1,
                                 zoom_range=0.1,
                                 horizontal_flip=True)
```

```python
train_generator=train_datagen.flow_from_directory('./Desktop/train/',
                                                 target_size=(224,224),
                                                 color_mode='rgb',
                                                 batch_size=32,
                                                 class_mode='categorical',
                                                 shuffle=True)
```

```python
step_size_train=train_generator.n//train_generator.batch_size
model.fit_generator(generator=train_generator,
                   steps_per_epoch=step_size_train,
                   epochs=10)
```

```python
img=image.load_img('./Desktop/ch.jpg',target_size=(224,224))
plt.imshow(img)
```

```python
x=image.img_to_array(img)
x=preprocess_input(x).reshape(1,224,224,3)
```
